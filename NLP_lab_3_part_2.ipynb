```python

def get_day(s):
    match = re.search(r'(\d{1,2})(?=[,t/snr])', s)
    if match:
        return int(match.group(1))
    return None
```

```python
def get_month(s):
  month_mapping = {
        "January": 1,
        "February": 2,
        "March": 3,
        "April": 4,
        "May": 5,
        "June": 6,
        "July": 7,
        "August": 8,
        "September": 9,
        "October": 10,
        "November": 11,
        "December": 12,
        "Jan":1,
        "Feb":2,
        "Mar": 3,
        "Apr": 4,
        "May": 5,
        "June": 6,
        "July": 7,
        "Aug": 8,
        "Sept": 9,
        "Oct": 10,
        "Nov": 11,
        "Dec": 12
    }

  month_reg= r'\b('+'|'.join(month_mapping.keys())+ r')\b'
  match=re.search(month_reg,s)
  if match:
    month_name=match.group(0)
    return month_mapping[month_name]
  return None
```

```python
def get_year(s):
    match = re.search(r'\b(\d{4})(?=\b|[.-/])', s)
    if match:
        return int(match.group(1))
    return None
```

```python
def gen_date(value):

    match = re.search(r'\b(\d{1,2})[./-](\d{1,2})[./-](\d{2})\b', value)
    if match:
        day, month, year = match.groups()
        return f"{day}/{month}/20{year}"

    match = re.search(r'\b(\d{1,2})[./-](\d{1,2})[./-](\d{4})\b', value)
    if match:
        day, month, year = match.groups()
        return f"{day}/{month}/{year}"

    match = re.search(r'(\d{4})[./-](\d{2})[./-](\d{2})', value)
    if match:
      year, month, day = match.groups()
      return f"{day}/{month}/{year}"

    date_1 = get_day(value)
    month = get_month(value)
    year = get_year(value)

    if date_1 and month and year:
        return f"{date_1}/{month}/{year}"

    return None
```

```python
df['Output']=df['Input'].apply(gen_date)
```
